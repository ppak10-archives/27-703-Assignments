```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
A = 1.9 * (10 ** -11)
m = 3.1

# Range of alternating stresses.
delta_sigma_min = 300
delta_sigma_max = 1200

# Range of number of cycles.
n_f_min = 10000
n_f_max = 10**7
n_f = np.arange(n_f_min, n_f_max, 10)

def delta_sigma(c_0):
    return ((1/(A * n_f * (m/2 - 1))) * c_0 ** (1 - m/2)) ** 1/m

# c_0 is provided in microns.
plt.plot(n_f, delta_sigma(10000000), color="black", label=f"{10000000/(10**7)} m")
plt.plot(n_f, delta_sigma(1000000), color="r", label=f"{1000000/(10**7)} m")
plt.plot(n_f, delta_sigma(100000), color="b", label=f"{100000/(10**7)} m")
plt.plot(n_f, delta_sigma(10000), color="g", label=f"{10000/(10**7)} m")

plt.legend()
plt.title("Initial Defect Size Contours")
plt.xlabel("Number of cycles in logscale (log(N sub f))")
plt.ylabel("Alternating Stress in MPa (Δσ)")
plt.xscale("log")
plt.xlim(n_f_min, n_f_max)
plt.ylim(delta_sigma_min, delta_sigma_max)
plt.show()
```
