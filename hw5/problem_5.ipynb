```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
A = 1.9 * (10 ** -11)
m = 3.1

# Range of alternating stresses.
delta_sigma_min = 300
delta_sigma_max = 1200

# Range of number of cycles.
n_f_min = 10000
n_f_max = 10**7
n_f = np.arange(n_f_min, n_f_max, 10)

def delta_sigma(c_0):
    return (((c_0 ** (1 - m/2))/(A * n_f * (m/2 - 1)))) ** (1/m)

# c_0 is provided in microns.
plt.plot(n_f, delta_sigma(10**-2), color=(0,1,0), label=f"{10**-2} m")
plt.plot(n_f, delta_sigma(10**-3), color=(1,0,0), label=f"{10**-3} m")
plt.plot(n_f, delta_sigma(10**-4), color=(0,0,0), label=f"{10**-4} m")
plt.plot(n_f, delta_sigma(10**-5), color=(1,0,1), label=f"{10**-5} m")
plt.plot(n_f, delta_sigma(10**-6), color=(0,1,1), label=f"{10**-6} m")
plt.plot(n_f, delta_sigma(10**-7), color=(1,1,0), label=f"{10**-7} m")
plt.plot(n_f, delta_sigma(10**-8), color=(0,0,1), label=f"{10**-8} m")

plt.legend()
plt.title("Initial Defect Size Contours")
plt.xlabel("Number of cycles in logscale (log(N sub f))")
plt.ylabel("Alternating Stress in MPa ()")
plt.xscale("log")
plt.xlim(n_f_min, n_f_max)
plt.ylim(delta_sigma_min, delta_sigma_max)
plt.show()
```
