```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# The different power levels in watts.
POWERS = [ 10, 50, 100, 300, 500 ]

# Range of speeds in m/s.
MIN_SPEED = 0.1 
MAX_SPEED = 1.4 
SPEEDS = np.arange(MIN_SPEED, MAX_SPEED, 0.01)
print(np.shape(SPEEDS))
```

```python
def cooling_rate(T_l, T_s, k, V, absorptivity, Q):
    return (2 * np.pi * k * T_l * T_s * V ) / ( absorptivity * Q)
```

```python
for power in POWERS:
    dT_dt = cooling_rate(582.5, 557, 110, SPEEDS, 0.32, power)
    plt.plot(SPEEDS, dT_dt, label=f"Power: {power} W")

plt.title("Cooling Rates of Al-10Si-1Mg around melting point")
plt.xlabel("Speed m/s")
plt.ylabel("Cooling Rate K/s")
plt.yscale("log")
plt.legend()
plt.show()   
```

```python
for power in POWERS:
    dT_dt = cooling_rate(1681, 1673, 6.7, SPEEDS, 0.48, power)
    plt.plot(SPEEDS, dT_dt, label=f"Power: {power} W")

plt.title("Cooling Rates of Ti-6Al-4V around melting point")
plt.xlabel("Speed m/s")
plt.ylabel("Cooling Rate K/s")
plt.yscale("log")
plt.legend()
plt.show()   
```
