```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_1_df = pd.read_csv("./csv/image_1_intercepts.csv")
counts, bins = np.histogram(image_1_df.iloc[:,6], bins=50)
```

```python
plt.title("Histogram of Image 1 Inverse Intercepts with 141 μm^2 area per point")
plt.ylabel("Counts")
plt.xlabel("Inverse Length (1 / μm)")
plt.hist(bins[:-1], bins, weights=counts)
plt.show()
```

```python
image_24_df = pd.read_csv("./csv/image_24_intercepts.csv")
counts, bins = np.histogram(image_24_df.iloc[:,6], bins=50)
```

```python
plt.title("Histogram of Image 24 Inverse Intercepts with 141 μm^2 area per point")
plt.ylabel("Counts")
plt.xlabel("Inverse Length (1 / μm)")
plt.hist(bins[:-1], bins, weights=counts)
plt.show()
```

```python
image_1_inverse_intercepts = image_1_df.iloc[:,6]
image_1_length = len(image_1_inverse_intercepts)
image_1_inverse_intercepts_sorted = np.sort(image_1_inverse_intercepts)
y = np.array(range(image_1_length))/image_1_length

plt.title("Cumulative Probability Plot for Image 1 Inverse Intercepts")
plt.xlabel("Inverse Length (1 / μm)")
plt.plot(image_1_inverse_intercepts_sorted, y)
plt.show()
```
