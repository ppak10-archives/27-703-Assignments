```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
image_1_df = pd.read_csv("./csv/image_1_intercepts.csv")
counts, bins = np.histogram(image_1_df.iloc[:,6], bins=50)
```

```python
plt.title("Histogram of Image 1 Intercepts with 141 μm^2 area per point")
plt.ylabel("Counts")
plt.xlabel("Length (μm)")
plt.hist(bins[:-1], bins, weights=counts)
plt.show()
```
