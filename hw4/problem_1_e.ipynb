```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.preprocessing import minmax_scale

# Records sample values length to truncate lists appropriately.
sample_1_length = 6
sample_4_length = 7
sample_5_length = 3
sample_6_length = 4
sample_7_length = 4
sample_8_length = 7
sample_9_length = 7
sample_10_length = 7
sample_12_length = 9

# Imports Powder values from csv file.
csv = pd.read_csv("./p1.csv")

# Imports all the relevant columns for each sample.
csv_sample_1_x = csv["Sample 1"]
csv_sample_1_y = csv["Unnamed: 3"]
csv_sample_4_x = csv["Sample 4"]
csv_sample_4_y = csv["Unnamed: 5"]
csv_sample_5_x = csv["Sample 4"]
csv_sample_5_y = csv["Unnamed: 7"]
csv_sample_6_x = csv["Sample 4"]
csv_sample_6_y = csv["Unnamed: 9"]
csv_sample_7_x = csv["Sample 7"]
csv_sample_7_y = csv["Unnamed: 11"]
csv_sample_8_x = csv["Sample 8"]
csv_sample_8_y = csv["Unnamed: 13"]
csv_sample_9_x = csv["Sample 9"]
csv_sample_9_y = csv["Unnamed: 15"]
csv_sample_10_x = csv["Sample 10"]
csv_sample_10_y = csv["Unnamed: 17"]
csv_sample_12_x = csv["Sample 12"]
csv_sample_12_y = csv["Unnamed: 19"]

# Starts at index 1 to remove the "X" and "Y" column headers.
csv_sample_1_x_data = csv_sample_1_x.iloc[1: sample_1_length] 
csv_sample_1_y_data = csv_sample_1_y.iloc[1: sample_1_length] 
csv_sample_4_x_data = csv_sample_4_x.iloc[1: sample_4_length] 
csv_sample_4_y_data = csv_sample_4_y.iloc[1: sample_4_length] 
csv_sample_5_x_data = csv_sample_5_x.iloc[1: sample_5_length] 
csv_sample_5_y_data = csv_sample_5_y.iloc[1: sample_5_length] 
csv_sample_6_x_data = csv_sample_6_x.iloc[1: sample_6_length] 
csv_sample_6_y_data = csv_sample_6_y.iloc[1: sample_6_length] 
csv_sample_7_x_data = csv_sample_7_x.iloc[1: sample_7_length] 
csv_sample_7_y_data = csv_sample_7_y.iloc[1: sample_7_length] 
csv_sample_8_x_data = csv_sample_8_x.iloc[1: sample_8_length] 
csv_sample_8_y_data = csv_sample_8_y.iloc[1: sample_8_length] 
csv_sample_9_x_data = csv_sample_9_x.iloc[1: sample_9_length] 
csv_sample_9_y_data = csv_sample_9_y.iloc[1: sample_9_length] 
csv_sample_10_x_data = csv_sample_10_x.iloc[1: sample_10_length]
csv_sample_10_y_data = csv_sample_10_y.iloc[1: sample_10_length]
csv_sample_12_x_data = csv_sample_12_x.iloc[1: sample_12_length]
csv_sample_12_y_data = csv_sample_12_y.iloc[1: sample_12_length]

# Converts values from string to float.
csv_sample_1_x_data = pd.to_numeric(csv_sample_1_x_data)
csv_sample_1_y_data = pd.to_numeric(csv_sample_1_y_data)
csv_sample_4_x_data = pd.to_numeric(csv_sample_4_x_data)
csv_sample_4_y_data = pd.to_numeric(csv_sample_4_y_data)
csv_sample_5_x_data = pd.to_numeric(csv_sample_5_x_data)
csv_sample_5_y_data = pd.to_numeric(csv_sample_5_y_data)
csv_sample_6_x_data = pd.to_numeric(csv_sample_6_x_data)
csv_sample_6_y_data = pd.to_numeric(csv_sample_6_y_data)
csv_sample_7_x_data = pd.to_numeric(csv_sample_7_x_data)
csv_sample_7_y_data = pd.to_numeric(csv_sample_7_y_data)
csv_sample_8_x_data = pd.to_numeric(csv_sample_8_x_data)
csv_sample_8_y_data = pd.to_numeric(csv_sample_8_y_data)
csv_sample_9_x_data = pd.to_numeric(csv_sample_9_x_data)
csv_sample_9_y_data = pd.to_numeric(csv_sample_9_y_data)
csv_sample_10_x_data = pd.to_numeric(csv_sample_10_x_data)
csv_sample_10_y_data = pd.to_numeric(csv_sample_10_y_data)
csv_sample_12_x_data = pd.to_numeric(csv_sample_12_x_data)
csv_sample_12_y_data = pd.to_numeric(csv_sample_12_y_data)

# Converts to numpy array.
csv_sample_1_x_data = csv_sample_1_x_data.to_numpy() 
csv_sample_1_y_data = csv_sample_1_y_data.to_numpy()
csv_sample_4_x_data = csv_sample_4_x_data.to_numpy()
csv_sample_4_y_data = csv_sample_4_y_data.to_numpy()
csv_sample_5_x_data = csv_sample_5_x_data.to_numpy()
csv_sample_5_y_data = csv_sample_5_y_data.to_numpy()
csv_sample_6_x_data = csv_sample_6_x_data.to_numpy()
csv_sample_6_y_data = csv_sample_6_y_data.to_numpy()
csv_sample_7_x_data = csv_sample_7_x_data.to_numpy()
csv_sample_7_y_data = csv_sample_7_y_data.to_numpy()
csv_sample_8_x_data = csv_sample_8_x_data.to_numpy()
csv_sample_8_y_data = csv_sample_8_y_data.to_numpy()
csv_sample_9_x_data = csv_sample_9_x_data.to_numpy()
csv_sample_9_y_data = csv_sample_9_y_data.to_numpy()
csv_sample_10_x_data = csv_sample_10_x_data.to_numpy()
csv_sample_10_y_data = csv_sample_10_y_data.to_numpy()
csv_sample_12_x_data = csv_sample_12_x_data.to_numpy()
csv_sample_12_y_data = csv_sample_12_y_data.to_numpy()

# Concatenates the data points into X and Y arrays.
X_data = np.concatenate([
    csv_sample_1_x_data,
    csv_sample_4_x_data,
    csv_sample_5_x_data,
    csv_sample_6_x_data,
    csv_sample_7_x_data,
    csv_sample_8_x_data,
    csv_sample_9_x_data,
    csv_sample_10_x_data,
    csv_sample_12_x_data,
])

Y_data = np.concatenate([
    csv_sample_1_y_data,
    csv_sample_4_y_data,
    csv_sample_5_y_data,
    csv_sample_6_y_data,
    csv_sample_7_y_data,
    csv_sample_8_y_data,
    csv_sample_9_y_data,
    csv_sample_10_y_data,
    csv_sample_12_y_data
])

# Scales values appropriately to their minimum and maximum. 
X_scaled = minmax_scale(X_data, (0, 10))
Y_scaled = minmax_scale(Y_data, (0, 100))

# Total values for x and y should be the same.
total_values = len(X_scaled)

# Calculates the rate parameter with scaled data.
rate_parameter = total_values / np.sum(X_scaled)

print('rate parameter:', rate_parameter)

# An array of A values determined from scaled X and Y values.
A = []

# Calculate the range of possible A values with scaled data.
for i, x in enumerate(X_scaled):
    A.append((Y_scaled[i] * np.exp(rate_parameter * x))/ rate_parameter)

# Set the bounds of A value to the maximum and minumum.
A_min = np.min(A)
A_max = np.max(A)

A_range = np.arange(A_min, A_max)

# Create a list of mean squared error values to find which A value best fits.
mean_squared_error_values = []

# Calculates error for A value.
for A_value in A_range:
    squared_error = (Y_scaled - A_value*rate_parameter*np.exp(-rate_parameter*X_scaled))**2
    mean_squared_error = np.sum(squared_error) / total_values
    mean_squared_error_values.append(mean_squared_error)

# Find the minimum mean squared error value.
mean_squared_error_min = np.min(mean_squared_error_values)

# Find the associated index.
mean_squared_error_min_index = np.array(mean_squared_error_values).argmin()

# Determines that the best fit A value is the one with the lowest error.
A_fit = A_range[mean_squared_error_min_index]
print(F"A: {A_fit}")

# Sort the X values to plot correctly
X_sort = np.sort(X_data)

# Calculates the appropriate Y values utilizing exponential PDF. 
Y_fit = A_fit*rate_parameter*np.exp(-rate_parameter*X_sort)

# Plot the Error when finding fitting A
plt.scatter(A_range, mean_squared_error_values, c="r", label="Mean Squared Error")
plt.title("A vs Error")
plt.xlabel("A")
plt.ylabel("Mean Squared Error")
plt.legend()
plt.show()

# Plot the data points along side the PDF with determined paramters (log scale).
plt.scatter(X_data, Y_data, c="b", label="Samples")
plt.plot(X_sort, Y_fit, c="g", label="Samples (fit)")
plt.yscale("log")
plt.title("Exponential PDF set to data (Log Scale)")
plt.xlabel("Spherical Equivalent Diameter (microns)")
plt.ylabel("Number of pores / mm^3")
plt.legend()
plt.show()


# Plot the data points along side the PDF with determined paramters.
plt.scatter(X_data, Y_data, c="b", label="Samples")
plt.plot(X_sort, Y_fit, c="g", label="Samples (fit)")
plt.title("Exponential PDF set to data")
plt.xlabel("Spherical Equivalent Diameter (microns)")
plt.ylabel("Number of pores / mm^3")
plt.legend()
plt.show()
```
